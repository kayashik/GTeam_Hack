```python
import pandas as pd
import hashlib
```

```python
with open('Ресурсы/Сессия 1/UserData.csv') as f:
    users = pd.read_csv(f, header=None)
users_ = users.copy()
users
```

## Prepare

```python
def to_md5(p):
    h = hashlib.md5(bytes(str(p), 'utf-8'))
    return h.hexdigest()

users[2] = users[2].apply(to_md5)
users
```

```python
users.loc[users[0] == 'Administrator', 0] = 1
users.loc[users[0] == 'User', 0] = 2
users
```

```python
users.loc[users[5] == 'Abu dhabi', 5] = 1
users.loc[users[5] == 'Cairo', 5] = 3
users.loc[users[5] == 'Bahrain', 5] = 4
users.loc[users[5] == 'Doha', 5] = 5
users.loc[users[5] == 'Riyadh', 5] = 6
users
```

```python
def transform_date(date):
    m, d, y = date.split('/')
    return '{}-{}-{}'.format(y, m, d)

users[6] = users[6].apply(transform_date)
users
```
