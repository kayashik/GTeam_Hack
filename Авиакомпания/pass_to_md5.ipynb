```python
import pandas as pd
import hashlib
```

```python
with open('Ресурсы/Сессия 1/UserData.csv') as f:
    users = pd.read_csv(f, header=None)
users
users_ = users.copy()
```

```python
def to_md5(p):
    h = hashlib.md5(bytes(str(p), 'utf-8'))
    return h.hexdigest()

users[2] = users[2].apply(to_md5)
users
```

```python
users.loc[users[0] == 'Administrator', 0] = 1
users.loc[users[0] == 'User', 0] = 2
users
```
